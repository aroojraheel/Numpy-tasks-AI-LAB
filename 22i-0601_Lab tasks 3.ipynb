# Lab Task 05

```python
print("           Welcome to estore:")
print("\n")
shopping_cart ={
    'Product': ['Cargo', 'Makeup', 'aqal', 'shame', 'life'],
    'Prices': [2, 3, 4, 5, 6]}
cart = []
flag=True
while flag:
    print(" GET A LIFE, nvm, Available products at our store are as follows:")
    for idx, product in enumerate(shopping_cart['Product'], 1):
        print(str(idx) + ". " + product + " - " + str(shopping_cart['Prices'][idx-1]))
    choice = input("Enter the number corresponding to the product you want to add to the cart (or 'quit' to quit): ")
    if choice.lower() == 'quit':
        break
    choice = int(choice)
    if choice < 1 or choice > len(shopping_cart['Product']):
        print("Invalid choice. Please enter a number within the range.")
        continue
    myprod = shopping_cart['Product'][choice - 1]
    selected_price = shopping_cart['Prices'][choice - 1]
    cart.append((myprod, selected_price))
    print(myprod,"has been added to your cart.")
print("\n")
print("Your Shopping Cart:")
total_price = 0
for item, price in cart:
    print(item,"-",price)
    total_price += price
print("\n")
print("Your total bill is:", total_price,"$")
print("Exitting you, bye.")
```


## Email Classification
Develope a tool which will classify an email.

### Tasks:
* Create a function classify_emails that takes a list of emails as an argument.
* Inside the function, categorize emails into "Important," "Promotions," and "Spam" based on keywords and content analysis.
* Loop through the emails and identify keywords to determine their category.
* If an email contains keywords indicating importance, assign it as "Important."
* If an email contains keywords related to promotions or deals, assign it as "Promotions."
* If an email is suspected to be spam based on certain keywords or patterns, assign it as "Spam."
* Return three lists of emails for each category.
* Print the lists.

```python
#code here
def classify(emails):
    ie = []
    pe = []
    se = []
    important_keywords = ['urgent', 'important', 'action required']
    promotion_keywords = ['promotion', 'deal', 'discount']
    spam_keywords = ['prize', 'claim now', 'free', 'click here']
    for email in emails:
        l= email.lower()

        if any(keyword in l for keyword in important_keywords):
            ie.append(email)
        elif any(keyword in l for keyword in promotion_keywords):
            pe.append(email)
        elif any(keyword in l for keyword in spam_keywords):
            se.append(email)
    return ie,pe,se
emails = ["Meeting today is very important","Congrats! you recieved a promotion","special offer for you","estore brings you exciting opportunities and deals","click here to earn a gift","you are awarded a prize for fre","urgent action required at place ofincident"]
important, promotions, spam = classify(emails)
print("Important emails are:")
for i in important:
  print(i)
print("\n")
print("Promotional emails are:")
for prm in promotions:
  print(prm)
print("\n")
print("Spam emails are: ")
for ss in spam:
  print(ss)
```

```python
#code here
spanish={
    'Thankyou':['Gracias'],
    'friend':['perra'],
    'girl':['chica'],
    'slap':['bofetada'],
    'okay':['bueno'],
    'hi':['ola']
}
flag=True
while flag:
  text=input("Enter an english word: ")
  x=text.lower()
  if x in spanish:
    print("Spanish translation of word is: ",spanish[x])
  elif x=='quit':
    print("Quitting our program")
    break
  else:
    print("Translation of the word is unavailable")
```
